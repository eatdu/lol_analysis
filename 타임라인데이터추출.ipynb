```python
import pandas as pd
import numpy as np
import requests
import json
import pickle
from tqdm import tqdm
```

```python
with open('timelineList3','rb') as f:
    timelineList3 = pickle.load(f)
len(timelineList3)
```

```python
plist = []
for index, status in enumerate(timelineList3):
    try :
        if status['metadata']['dataVersion'] == 2:
            plist.append(index)
        else : pass
    except: pass
print(plist)
```

```python
timeline, give_something, summonerData = [],[],[]
for tList in tqdm(range(len(timelineList3))):
    try:
        for i in range(len(timelineList3[tList]['info']['frames'])):
            for v in timelineList3[tList]['info']['frames'][i]['events']:
                x,y = 0,0
                if v.get('position'):
                    x = v.get('position')['x']
                    y = v.get('position')['y']
                timeline.append([timelineList3[tList]['info']['gameId'],v.get('realTimestamp',0),v.get('itemId',0),v.get('level',0)
                                    ,v.get('levelUpType',0),v.get('creatorId',0),v.get('participantId',0),v.get('timestamp',0)
                                    ,v.get('type',0),v.get('skillSlot',0),v.get('monsterType',0),v.get('wardType',0)
                                    ,v.get('killType',0),v.get('killerId',0),v.get('victimId',0), x,y,v.get('winningTeam',0)])
            for num in range(10):
                give_something.append([timelineList3[tList]['info']['gameId'],timelineList3[tList]['info']['frames'][i]['timestamp'],
                                       timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['participantId']
                                          , timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['totalGold'],
                                       timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['level']
                                          , timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)][
                                           'jungleMinionsKilled'] +
                                       timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['minionsKilled']
                                          , timelineList3[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['xp']
                                       ])
        for userId in range(10):
            summonerData.append([timelineList3[tList]['info']['gameId'],timelineList3[tList]['info']['participants'][userId]['participantId']
                                 ,timelineList3[tList]['info']['participants'][userId]['puuid']])
    except:
        pass

timelineData = pd.DataFrame(timeline, columns=['gameId','realTimestamp','itemId','level','levelUpType'
    ,'creatorId','participantId','timestamp','type','skillSlot','monsterType','wardType','killType','killerId'
    ,'victimId','position_x','position_y','winningTeam'])
timePerChampionData = pd.DataFrame(give_something, columns=['gameId','timestamp','participantId','totalGold','level','CS','xp'])
summonerData = pd.DataFrame(summonerData, columns=['gameId','participantId','puuid'])
```

```python
with open('timelineData3','wb') as f:
    pickle.dump(timelineData,f)
with open('timePerChampionData3','wb') as f:
    pickle.dump(timePerChampionData,f)
with open('summonerData3','wb') as f:
    pickle.dump(summonerData,f)
```
