```python
timelineList = []
api_key = 'RGAPI-1375b084-a7cf-4807-9077-820a26384391'
for gameId in tqdm(gameIdList):
    url = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + gameId + '/timeline?api_key=' + api_key
    datas = requests.get(url)
    datas = datas.json()
    time.sleep(1.7)
    timelineList.append(datas)
len(timelineList)
```

```python
with open('timelineList.pickle','wb') as f:
    pickle.dump(timelineList, f)
```

```python
timelineList[0]
```

```python
timeline, give_something, summonerData = [],[],[]
for tList in tqdm(range(len(timelineList))):
    try:
        for i in range(len(timelineList[tList]['info']['frames'])):
            for v in timelineList[tList]['info']['frames'][i]['events']:
                x,y = 0,0
                if v.get('position'):
                    x = v.get('position')['x']
                    y = v.get('position')['y']
                timeline.append([timelineList[tList]['info']['gameId'],v.get('realTimestamp',0),v.get('itemId',0),v.get('level',0)
                                    ,v.get('levelUpType',0),v.get('creatorId',0),v.get('participantId',0),v.get('timestamp',0)
                                    ,v.get('type',0),v.get('skillSlot',0),v.get('monsterType',0),v.get('wardType',0)
                                    ,v.get('killType',0),v.get('killerId',0),v.get('victimId',0), x,y,v.get('winningTeam',0)])
            for num in range(10):
                give_something.append([timelineList[tList]['info']['gameId'],timelineList[tList]['info']['frames'][i]['timestamp'],
                                       timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['participantId']
                                          , timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['totalGold'],
                                       timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['level']
                                          , timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)][
                                           'jungleMinionsKilled'] +
                                       timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['minionsKilled']
                                          , timelineList[tList]['info']['frames'][i]['participantFrames'][str(num + 1)]['xp']
                                       ])
        for userId in range(10):
            summonerData.append([timelineList[tList]['info']['gameId'],timelineList[tList]['info']['participants'][userId]['participantId']
                                 ,timelineList[tList]['info']['participants'][userId]['puuid']])
    except:
        pass

timelineData = pd.DataFrame(timeline, columns=['gameId','realTimestamp','itemId','level','levelUpType'
    ,'creatorId','participantId','timestamp','type','skillSlot','monsterType','wardType','killType','killerId'
    ,'victimId','position_x','position_y','winningTeam'])
timePerChampionData = pd.DataFrame(give_something, columns=['gameId','timestamp','participantId','totalGold','level','CS','xp'])
summonerData = pd.DataFrame(summonerData, columns=['gameId','participantId','puuid'])
```
